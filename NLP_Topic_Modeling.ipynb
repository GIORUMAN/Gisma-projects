```python
df = pd.read_csv(r'C:\Users\USER\Downloads\tweets.csv')
df
```

lowrcasing

```python
df = df.apply(lambda x: x.astype(str).str.lower())
text_col = df['text']
print(df['text'])
```

tokenization

```python
text_col = 'text'
df[text_col] = df[text_col].str.replace(r'[^\w\s]', '', regex=True) 
tokens = df[text_col].apply(lambda x: word_tokenize(x))
print(tokens)
```

removing tags/handles 

```python
pattern = r"@[^\s]+|#[^\s]+"
df[text_col] = df[text_col].apply(lambda x: re.sub(pattern, '', x))
text = text_col
print(df[text])
```

drop stop words 

```python
stop_words = stopwords.words('english')
def remove_stopwords(text):
  tokens = word_tokenize(text)
  filtered_words = [word for word in tokens if word not in stop_words]
  return ' '.join(filtered_words)
df[text_col] = df[text_col].apply(lambda x: ' '.join([word for word in word_tokenize(x) if word not in stop_words]))
text = text_col
print(df[text])
```

Removing puctuations 

```python
df[text_col] = df[text_col].str.translate(str.maketrans('', '', '!@#$%^&*()-=+{};:\'<,>.?/\\|`~"[\]'))
text = df[text_col] 
print(df[text_col].head(14640))
```

```python
nltk_lemmatizer = WordNetLemmatizer()
spacy_lemmatizer = spacy.load('en_core_web_sm')
def get_wordnet_pos(word):
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}
    return tag_dict.get(tag, wordnet.NOUN)
def nltk_lemmatize(text):
    tokens = word_tokenize(text)
    lemmatized_tokens = [nltk_lemmatizer.lemmatize(token, get_wordnet_pos(token)) for token in tokens]
    return ' '.join(lemmatized_tokens)
    
print(df[text_col].head(14640))
```
